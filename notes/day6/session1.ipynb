# Linear Regression

Linear regression is a type of supervised machine-learning algorithm that learns from the labelled datasets and maps the data points with most optimized linear functions which can be used for prediction on new datasets.

### working of linear regression 
- the straight line equation is y=mx+b
- the regression model will calculate the values of coefficient (m) and intercept (b)
- so when we try to predict a value it will substitute it and predict the value 

---

## Line of Best fit concept

In linear regression, the best-fit line is the straight line that most accurately represents the relationship between the independent variable (input) and the dependent variable (output). It is the line that minimizes the difference between the actual data points and the predicted values from the model.
It is the straight line that most closely follows the pattern of the data points.

### Why it is called Best fit 
Because it is chosen such that the total error between the line and the data points is minimized.
 
### How line of Best fit is chosen 
- Pick a straight line shape
- Assume the line looks like: ùë¶=ùëöùë•+ùëê
- Calculate the predicted value for each data point.
- Find the error :Difference between actual value and predicted value.
- Square all the errors (So negative errors don‚Äôt cancel out.)
- Add all the squared errors This gives the total error of that line.
- Choose the line with the smallest total error
- This is the line of best fit. <br><br>

```python
# Import the required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# read the csv 
df = pd.read_csv('./resources/homeprices.csv')
df.head()
```

```python
# plot a scatter plot
plt.xlabel('area')
plt.xlabel('price')
plt.scatter(df.area , df.price , color='red' , marker='+')
```

```python
# implementing the linear regression 
reg = linear_model.LinearRegression()
reg.fit(df[['area']] , df.price)
```

```python
# predicting a value
reg.predict([[3300]])
```

```python
# printing the calculated coefficient and intercept
print(reg.coef_)
print(reg.intercept_)
```

```python
# scatter plot along with best fit line 
plt.xlabel('area')
plt.xlabel('price')
plt.scatter(df.area , df.price , color='red' , marker='+')
plt.plot(df.area , reg.predict(df[['area']])  )
```
