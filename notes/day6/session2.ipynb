# Correlation vs Causation

## 1. Correlation
**Correlation** means two variables are related or move together.  
When one changes, the other also tends to change — **but this does NOT mean one causes the other.**

### Types of Correlation
- **Positive correlation**: Both variables increase or decrease together  
  *Example:* Height ↑ and Weight ↑  
- **Negative correlation**: One increases while the other decreases  
  *Example:* Speed ↑ and Travel Time ↓  
- **Zero correlation**: No meaningful relationship

### Example
- Ice cream sales ↑  
- Drowning incidents ↑  

Both rise during summer because of **hot weather** — but ice cream does *not* cause drowning.

---

## 2. Causation
**Causation** means one variable *directly causes* a change in another.

### Example
- Smoking → causes → Lung cancer  
This is a proven cause-and-effect relationship.

---

## Why Correlation ≠ Causation
Just because two things move together does not mean one causes the other.

Reasons:
1. **Coincidence**
2. **Hidden third factor (confounder)**
3. **Reverse causality** (B causes A, not A causes B)

### Example of a Confounder
- More firefighters at a fire → More damage  
People might think firefighters cause damage, but:  
**Bigger fires → need more firefighters → cause more damage**  
The **size of the fire** is the hidden factor.

---

## Summary
- **Correlation** = relationship  
- **Causation** = cause-and-effect  
- Correlation alone cannot prove causation  
- Causation requires **controlled experiments, evidence, and analysis**


## Implementation in python

### Correlation
- You can use pandas and numpy to find correlation in python

```python
import pandas as pd

data = {
    "hours_studied": [1, 2, 3, 4, 5, 6],
    "marks_scored":  [40, 50, 60, 70, 80, 90],
    "ice_cream_sales": [100, 200, 300, 400, 500, 600]
}

df = pd.DataFrame(data)

# A : correlation between all columns
corr_matrix = df.corr()
print(corr_matrix)

# B : correlation between two specific columns
corr = df['hours_studied'].corr(df['marks_scored'])
print("Correlation of two columns:", corr)

# C. Visualize correlation using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True)
plt.show()
```

### Causation  

- You cannot directly find Causation in Python.
- However, you can test causation using statistical models like Linear Regression

```python
# Checking if studying increases marks
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array(df['hours_studied']).reshape(-1, 1)
y = np.array(df['marks_scored'])

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("R² Score:", model.score(X, y))

# Slope > 0 → hours studied increases marks
# High R² → strong causal-like relationship
```
