# Logistic Regression

Logistic Regression is a supervised machine learning algorithm used for classification problems. Unlike linear regression which predicts continuous values it predicts the probability that an input belongs to a specific class.

## Types of Logistic Regression

1. Binary Logistic Regression: Used when the target variable has two classes (e.g., yes/no, 0/1).
2. Multinomial Logistic Regression: Used when the target variable has more than two classes without any order (e.g., cat/dog/mouse).
3. Ordinal Logistic Regression: Used when the target variable has more than two classes with a specific order (e.g., low/medium/high).

## Binary Classification concept and Sigmoid Function

## Binary Classification
Binary Classification

Binary classification is a type of machine learning problem where the goal is to predict one of two possible outcomes. In the context of supervised learning, you are typically given a dataset with input features (or independent variables), and the task is to predict a binary outcome (or dependent variable), which can take on two distinct classes or categories.

For example:

- Spam email detection: The classes could be "Spam" or "Not Spam".

- Medical diagnosis: The classes could be "Disease" or "No Disease".

- Credit card fraud detection: The classes could be "Fraudulent" or "Not Fraudulent".

In binary classification, the model learns from the training data and attempts to assign a label of either 0 or 1 (or any two classes of your choice) to each input.

## Sigmoid Function

The sigmoid function, also known as the logistic function, is crucial in logistic regression. It maps the output of the linear model (a real number) into a probability that lies between 0 and 1. This is important because the task of binary classification is to predict the likelihood of a certain class, and probabilities naturally fit this scenario.
The formula for the sigmoid function is:

```
S(x) = 1 / (1 + e^(-x))
```

````python
# Predicting if a person would buy life insurnace based on his age using logistic regression
import pandas as pd
from matplotlib import pyplot as plt
````

````python
df = pd.read_csv("./resources/insurance_data.csv")
df.head()
````

````python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.9)


from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)

print(X_test)
model.predict(X_test)
model.score(X_test,y_test)

prob = model.predict_proba(X_test)

print("Probability of not buying and buying insurance respectively for the test data:\n")
pd.DataFrame(prob,columns = ['Not Buy','Buy'])
````
