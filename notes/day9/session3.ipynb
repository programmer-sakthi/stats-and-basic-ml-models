# K - means clustering 

K-means clustering is an unsupervised machine learning algorithm used to group similar data points into clusters based on their features. The algorithm works by partitioning the data into K clusters, where K is a user-defined parameter. The goal of K-means clustering is to minimize the distance between data points within the same cluster while maximizing the distance between different clusters.


## components of K - means clustering 

### Centeroid :
In K-Means clustering, a centroid is the center point of a cluster.
It represents the mean position of all the data points assigned to that cluster.

- Each cluster has one centroid.

- Centroids are not actual data points; they are calculated values.

- During training, K-Means moves centroids repeatedly to minimize the within-cluster distances.

- The goal is to place centroids such that the total distance between each point and its cluster’s centroid (inertia) is minimized.

### Elbow Method : 

The Elbow Method is a technique to determine the optimal number of clusters (K) to use in K-Means.
Choosing too few clusters oversimplifies the pattern, while too many clusters overfits the data.
The elbow method helps find a balance.

#### How it works ? 

- Run K-Means for different values of K (e.g., 1 to 10).

- For each K, compute inertia — the sum of squared distances between each point and its cluster centroid.

- Plot K vs. inertia.

##### Interpretation 

- As K increases, inertia decreases (clusters fit better).

- But after a certain K, the rate of improvement drops sharply.

- This point on the graph where the curve bends like an “elbow” is considered the optimal K.
