Cross-validation is a resampling technique used to evaluate the performance of a model on unseen data.
It reduces overfitting, improves generalization, and provides a more reliable estimate of model performance.

| Technique         | Best For                    | Pros                              | Cons              |
| ----------------- | --------------------------- | --------------------------------- | ----------------- |
| Hold-Out          | Quick baseline              | Fast                              | High variance     |
| K-Fold            | General ML                  | Balanced bias & variance          | Slower            |
| Stratified K-Fold | Classification (imbalanced) | Maintains distribution            | Slightly complex  |
| LOOCV             | Small datasets              | Almost all data used for training | Very slow         |
| Leave-P-Out       | Research                    | Accurate                          | Too expensive     |
| Repeated K-Fold   | Reliable evaluation         | Less variance                     | Slow              |
| Nested CV         | Hyperparameter tuning       | Very reliable                     | Very expensive    |
| Time Series CV    | Forecasting                 | True time structure               | Specific use only |
