# Accuracy, Precision, Recall & F1-Score  
### Model Evaluation ‚Äì Machine Learning

---

## **1Ô∏è‚É£ Accuracy**
Accuracy measures the overall correctness of the model.

### **Formula**
$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

### **When to Use**
- Works well when classes are **balanced**
- Misleading with **imbalanced datasets**

---

## **2Ô∏è‚É£ Precision**
Out of all predicted positive cases, how many are actually positive?

### **Formula**
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

### **Meaning**
- High precision ‚Üí fewer **false positives**
- Useful in spam detection, fraud detection

---

## **3Ô∏è‚É£ Recall (Sensitivity)**
Out of all actual positives, how many did the model correctly identify?

### **Formula**
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

### **Meaning**
- High recall ‚Üí fewer **false negatives**
- Useful for medical tests, safety-critical detection

---

## **4Ô∏è‚É£ F1-Score**
Harmonic mean of Precision and Recall.

### **Formula**
$$
F1 = 2 \times \frac{(\text{Precision} \times \text{Recall})}{(\text{Precision} + \text{Recall})}
$$

### **Meaning**
- High only when both **Precision & Recall** are high
- Best for **imbalanced datasets**

---

## üìå Summary Table

| Metric | Formula | Best Used When | Avoid When |
|--------|---------|----------------|-------------|
| **Accuracy** | $$\frac{TP+TN}{TP+TN+FP+FN}$$ | Balanced classes | Imbalanced datasets |
| **Precision** | $$\frac{TP}{TP+FP}$$ | Minimize FP | FN cost is higher |
| **Recall** | $$\frac{TP}{TP+FN}$$ | Minimize FN | FP cost is higher |
| **F1-Score** | $$2 \times \frac{PR}{P + R}$$ | Balance needed | Accuracy is enough |

## ‚≠ê Example (Optional)

Given:

- TP = 80  
- FP = 20  
- FN = 40  
- TN = 860  

### Computed Metrics

**Accuracy**  
$$
\frac{80 + 860}{1000} = 0.94
$$

**Precision**  
$$
\frac{80}{80 + 20} = 0.80
$$

**Recall**  
$$
\frac{80}{80 + 40} = 0.667
$$

**F1 Score**  
$$
2 \times \frac{0.80 \times 0.667}{0.80 + 0.667} = 0.727
$$
