```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv
df=pd.read_csv('./ML374_S5_Practice_Income_Data.csv')
df.head()
columns=df.columns.tolist()
category_columns = [
    "WorkClass",
    "Education",
    "Occupation",
    "Relationship",
    "Gender",
    "Native_Country"
]

# Convert to category dtype
for col in category_columns:
    df[col] = df[col].astype("category")
# print(df.dtypes)
```

```python
categorical_cols = df.select_dtypes(include=['category']).columns.tolist()
```

```python
# define the subplot dimensions 
num_plots = len(categorical_cols)
num_cols = 3        # fixed number of columns in the grid
num_rows = (num_plots + num_cols - 1) // num_cols
```

```python
plt.figure(figsize=(20, 5 * num_rows))
for i, col in enumerate(categorical_cols, 1):

    freq = df[col].value_counts()

    # Create custom labels: value_<category>
    labels = [f"value_{str(v)}" for v in freq.index]

    plt.subplot(num_rows, num_cols, i)
    plt.bar(labels, freq.values)
    plt.title(col)
    plt.xlabel("Category")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
