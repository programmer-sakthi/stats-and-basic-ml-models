```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load file
df = pd.read_csv('./ML374_S6_Concept_Weather_Cleaned_Data.csv') 
df.head()
```

```python
# Split data into training and testing sets
X = df[['global_radiation']]
y = df['temperature']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=200)
```

```python
# Model
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
residuals = y_test - y_pred
```

```python
# ===========================
# 1. Scatter: actual data + regression line
# ===========================
plt.figure(figsize=(6,6))
plt.scatter(df['global_radiation'], df['temperature'], alpha=0.6)
plt.plot(X, model.predict(X), color='yellow')      # regression line
plt.xlabel('global_radiation')
plt.ylabel('temperature')
plt.title('')
plt.show()
```

```python
# ===========================
# 2. Actual vs Predicted
# ===========================
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'y--')  # diagonal
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted')
plt.show()
```

```python
# ===========================
# 3. Residual Plot vs Index
# ===========================
plt.figure(figsize=(6,6))
plt.scatter(range(len(residuals)), residuals, alpha=0.6)
plt.axhline(0, color='orange', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residual plot')
plt.show()
```

```python
# ===========================
# 4. Histogram of Errors
# ===========================
plt.figure(figsize=(6,5))
plt.hist(residuals, bins=20, density=True, alpha=0.7)
plt.xlabel('Error Value')
plt.ylabel('Probability Density')
plt.title('Errors Distribution')
plt.show()
```

```python
# ===========================
# 5. Regression line again (bottom final)
# ===========================
plt.figure(figsize=(6,6))
plt.scatter(df['global_radiation'], df['temperature'], alpha=0.6)
plt.plot(X, model.predict(X), color='yellow')
plt.xlabel('global_radiation')
plt.ylabel('temperature')
plt.show()
```
